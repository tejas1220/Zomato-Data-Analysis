# Required Library and dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv(r"C:\Users\tejas\Downloads\Zomato data .csv")
print(dataframe.head())
```

```python
dataframe.info()
```

# converting data type of column- rate

```python
def handleRate(value):
    return float(str(value).split('/')[0])

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

# Type of Resturant

```python
 dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of Resturant")
```

# Conclusion :- Majority of the resturant falls in dinning category

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

# Conclusion - Dinning resturant receive maximum votes

# Rating give by people to resturant

```python
 dataframe.head()
```

```python
plt.hist(dataframe['rate'], bins=5, rwidth=0.8)
plt.title("Rating distribution")
plt.show()
```

 # conclusion - The majority resturants received ratings from 3.50 to 4.00

# *Average order :-

```python
 dataframe.head()
```

```python
avg_order = dataframe['Average_order']
sns.countplot(x=avg_order)
```

# Conclusion - The majority of people prefer resturants with average order is 300 rupees

# * Which mode receives maximum rating

```python
 dataframe.head()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='online_order', y='rate', data=dataframe)
```

# Conclusion - Offline order received lower rating in camparison to online order

# * “Online vs Offline Orders Across Restaurant Types”

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

# Conclusion - Dining resturants primarily accept offline orders, where cafes  primarily receive online orders. This suggests that clients preference orders in person at resturants, but prefer online ordering at cafes
